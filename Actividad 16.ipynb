### Crea una gráfica de barras que represente a los 10 países con mas medallas ganadas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leer el archivo csv, en la instruccion dependiendo el tipo de archivo es que se le cambia la instrucción de read
datos1=pd.read_csv("C:\\Users\\PALDAPE\\Downloads\\athlete_events.csv")
```

```python
datos1
```

```python
#sacar solo a los paises que han ganado medalla
ganadores=datos1[datos1["Medal"].notna()]
```

```python
ganadores
```

```python
dicc=dict(ganadores["NOC"].value_counts())
```

```python
dicc
```

```python
#obtener el top 10
top_10= dict(list(dicc.items())[0: 10])
```

```python
top_10
```

```python
#graficar
paises = list(top_10.keys())
medallas = list(top_10.values())

plt.bar(paises, medallas)
plt.show()
```

## AVANZADO

```python
sorted_dicc={}
sorted_keys = sorted(top_10, key=top_10.get) 
sorted_keys
```

```python
for w in sorted_keys:
    sorted_dicc[w] = top_10[w]

print(sorted_dicc)
```

```python
#graficar
paises = list(sorted_dicc.keys())
medallas = list(sorted_dicc.values())

def labels(paises,medallas):
    for i in range(len(paises)):
        plt.text(i,medallas[i],medallas[i]) 
```

```python
plt.figure(figsize=(10,5))
plt.bar(paises,medallas)
labels(paises,medallas)
plt.show
```
