### Crea y guarda una grafica de pay del año mas reciente de la base de datos

```python
import pandas as pd
import numpy as np
import matplotlib as plot
```

```python
#leer el archivo csv, en la instruccion dependiendo el tipo de archivo es que se le cambia la instrucción de read
datos1=pd.read_csv("C:\\Users\\PALDAPE\\Downloads\\athlete_events.csv")
```

```python
datos1
```

```python
año_max=datos1["Year"].max()
```

```python
año_max
```

```python
datos_recientes=datos1[datos1["Year"]==año_max]
```

```python
datos_recientes
```

```python
dummies=pd.get_dummies(datos_recientes[["Medal"]])
dummies.head()
```

```python
#juntar ambos dataframes
concatenado=pd.concat([datos_recientes,dummies],axis=1)
concatenado.head()
```

```python
agrupado=concatenado.groupby(["Year","Medal","Season"]).sum()
agrupado.head(8)
```

```python
agrupado["Total_Medals"]=agrupado["Medal_Gold"]+agrupado["Medal_Silver"]+agrupado["Medal_Bronze"]
```

```python
agrupado
```

```python
cantidad=agrupado["Total_Medals"][:3].unique()
cantidad
```

```python
agrupado["Total_Medals"][:3]
```

```python
#incluirle los nombres de las categorias
nombres=["Bronce", "Oro", "Plata"]
```

```python
pie2=agrupado["Total_Medals"][:3].plot.pie(figsize=(10,10), labels=nombres, autopct="%0.1f %%", title="Medallas ganadas en el 2016")
```

```python
#guardar el grafico
fig=pie2.get_figure()
fig.savefig("Gráfica de pay")
```
